```python
import pandas as pd
from evaluation_tools import get_metrics, print_metrics


# random is a randomly sampled distribution, not compressed.
def datagetter_opensub(lang1, lang2, split="test", comp="random"):
    return pd.read_csv(
        f"data/opensubtitles/compressed/{lang1}-{lang2}-{comp}.{split}.csv"
    )


metrics, translations = get_metrics(
    languages=["fr", "lt", "hu", "pl"],
    datagetter=datagetter_opensub,
    device="cuda:0"
)
```

```python
for lang, metric in metrics.items():
    print_metrics(lang, metric)
```
